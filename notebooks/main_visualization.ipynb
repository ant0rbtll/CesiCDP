# Fichier `main_visualization.py`

Ce notebook explique le role du lanceur `main_visualization.py`. Ce fichier n'est pas une brique metier du solveur : c'est un point d'entree pratique pour ouvrir une fenetre `matplotlib` classique et tester le bouton `->` hors notebook.

## Pourquoi un lanceur hors notebook ?

Les boutons `matplotlib` peuvent etre instables selon le frontend Jupyter utilise. Le script `main_visualization.py` ouvre une vraie fenetre graphique, ce qui rend l'interaction plus fiable.

Il construit :

- une `GraphGenerationConfig` a partir des arguments de ligne de commande ;
- un `GraphGenerator` ;
- une `GraphInstance` ;
- un `GraphVisualizer` ;
- puis il appelle `plt.show()` pour garder la fenetre ouverte.

## Commande minimale

Depuis la racine du depot :

```bash
python3 main_visualization.py
```

Tu peux ensuite cliquer sur `->` pour faire avancer la dynamique du graphe.

## Parametres utiles

Le script expose plusieurs options :

- `--nodes` : nombre de sommets ;
- `--seed` : scenario aleatoire reproductible ;
- `--sigma` : amplitude des variations dynamiques ;
- `--mean-reversion` : force de rappel vers le cout statique ;
- `--max-multiplier` : borne haute du cout dynamique ;
- `--forbid-prob` : probabilite de passage temporaire en `OFF` ;
- `--restore-prob` : probabilite de retour a la normale ;
- `--max-disabled-ratio` : part maximale de routes dynamiquement indisponibles.

Exemple :

```bash
python3 main_visualization.py --nodes 12 --seed 7 --forbid-prob 0.02
```
