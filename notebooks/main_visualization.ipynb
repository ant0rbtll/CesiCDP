# Fichier `main_visualization.py`

Ce notebook documente le lanceur `main_visualization.py`. Il n'est pas une brique metier : c'est un point d'entree pratique pour ouvrir une fenetre `matplotlib` classique avec un vrai backend GUI, ou le bouton `->` est fiable.

## Pourquoi un lanceur hors notebook ?

Les widgets matplotlib (`Button`, `Slider`, etc.) dependent du backend. Avec `%matplotlib inline` (par defaut dans beaucoup de frontends Jupyter), les images sont statiques et les clics ne sont pas transmis.

`main_visualization.py` ouvre une fenetre native (`plt.show()`) qui reagit correctement au bouton. Cela facilite les demos pedagogiques.

## Ce que le script fait

1. Parse les arguments de ligne de commande (voir section parametres).
2. Construit une `GraphGenerationConfig` correspondante.
3. Appelle `GraphGenerator(config).generate()`.
4. Instancie un `GraphVisualizer` et appelle `show_dynamic_graph()`.
5. Laisse la fenetre ouverte avec `plt.show()`.

## Commande minimale

Depuis la racine du depot :

```bash
python3 main_visualization.py
```

Le clic sur `->` fait avancer la dynamique d'un tour.

## Parametres disponibles

Tous optionnels. Valeurs par defaut dans `GraphGenerationConfig`.

| Option | Parametre config | Exemple |
|---|---|---|
| `--nodes` | `node_count` | `--nodes 12` |
| `--seed` | `seed` | `--seed 7` |
| `--sigma` | `dynamic_sigma` | `--sigma 0.2` |
| `--mean-reversion` | `dynamic_mean_reversion_strength` | `--mean-reversion 0.4` |
| `--max-multiplier` | `dynamic_max_multiplier` | `--max-multiplier 2.0` |
| `--forbid-prob` | `dynamic_forbid_probability` | `--forbid-prob 0.05` |
| `--restore-prob` | `dynamic_restore_probability` | `--restore-prob 0.3` |
| `--max-disabled-ratio` | `dynamic_max_disabled_ratio` | `--max-disabled-ratio 0.25` |

Exemple combine :

```bash
python3 main_visualization.py --nodes 12 --seed 7 --forbid-prob 0.05 --sigma 0.2
```

## Lien avec le notebook `visualization.ipynb`

Le notebook et le lanceur utilisent la **meme** classe `GraphVisualizer`. La seule difference est le backend matplotlib :

- notebook : souvent `inline`, vue statique, bouton via `advance_session` ;
- lanceur : fenetre native, bouton reactif, ideal pour demos orales.
