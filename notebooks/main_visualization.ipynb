# Fichier `main_visualization.py`

Ce notebook documente le lanceur `main_visualization.py`. Il n'est pas une brique metier : c'est un point d'entree pratique pour ouvrir une fenetre `matplotlib` classique avec un vrai backend GUI [1, 2], ou le bouton `->` est plus fiable que dans beaucoup de notebooks.

Le fichier est volontairement petit. Il assemble des briques deja expliquees ailleurs : `GraphGenerationConfig`, `GraphGenerator`, `GraphVisualizer` et `DynamicNetworkSimulator` via la session interactive.


## Pourquoi un lanceur hors notebook ?

Les widgets matplotlib (`Button`, `Slider`, etc.) dependent du backend [2, 3]. Avec `%matplotlib inline`, les images sont souvent statiques et les clics ne sont pas transmis a Python.

`main_visualization.py` ouvre une fenetre native via `plt.show()`. Dans ce mode, le bouton `->` appelle vraiment le callback, le snapshot avance, et la figure se met a jour.

Le notebook `visualization.ipynb` reste utile pour expliquer les vues et inspecter les objets. Le script est meilleur pour manipuler la simulation comme une petite application locale.


## Ce que le script fait

1. Parse les arguments de ligne de commande.
2. Construit une `GraphGenerationConfig` avec les valeurs demandees.
3. Genere une instance valide via `GraphGenerator(config).generate()`.
4. Affiche un resume textuel de l'instance dans le terminal.
5. Instancie un `GraphVisualizer`.
6. Lance `show_dynamic_graph()`.
7. Bloque sur `plt.show()` pour garder la fenetre ouverte.

Le script ne contient pas de logique de simulation propre. Toute la dynamique reste dans les modules `dynamic_network.py`, `dynamic_costs.py` et `visualization.py`.


## Commande minimale

Depuis la racine du depot :

```bash
python3 main_visualization.py
```

Le clic sur `->` fait avancer la dynamique d'un tour. A chaque clic, les couts affiches peuvent changer et certaines routes peuvent passer temporairement OFF, tant que les validateurs acceptent l'etat obtenu.

Si la fenetre ne s'ouvre pas, le probleme vient generalement du backend graphique disponible sur la machine, pas du generateur lui-meme. Dans ce cas, le notebook `visualization.ipynb` permet encore d'avancer manuellement la session.


## Parametres disponibles

Tous les parametres sont optionnels. Les valeurs par defaut restent proches de `GraphGenerationConfig`.

| Option | Parametre config | Exemple |
|---|---|---|
| `--nodes` | `node_count` | `--nodes 12` |
| `--seed` | `seed` | `--seed 7` |
| `--sigma` | `dynamic_sigma` | `--sigma 0.2` |
| `--mean-reversion` | `dynamic_mean_reversion_strength` | `--mean-reversion 0.4` |
| `--max-multiplier` | `dynamic_max_multiplier` | `--max-multiplier 2.0` |
| `--forbid-prob` | `dynamic_forbid_probability` | `--forbid-prob 0.05` |
| `--restore-prob` | `dynamic_restore_probability` | `--restore-prob 0.25` |
| `--max-disabled-ratio` | `dynamic_max_disabled_ratio` | `--max-disabled-ratio 0.3` |

Les options les plus sensibles sont `--forbid-prob` et `--max-disabled-ratio`. La premiere augmente les tentatives de coupure, la seconde limite ce que le validateur laissera vraiment passer.


## Lien avec le notebook `visualization.ipynb`

Le notebook et le lanceur utilisent la **meme** classe `GraphVisualizer`. La difference est le contexte d'execution :

- notebook : pratique pour expliquer, capturer des figures et inspecter `session.snapshot` ;
- script : pratique pour cliquer dans une fenetre interactive stable.

Cette complementarite evite de surcharger le notebook avec des details de backend GUI. Le code pedagogique reste lisible, et le script offre une experience plus proche d'un outil utilisateur.

A retenir : `main_visualization.py` est un lanceur, pas une deuxieme implementation. Si le comportement visuel doit changer, il faut modifier `GraphVisualizer`, pas dupliquer la logique dans le script.
