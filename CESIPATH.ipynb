# Table des Figures

```python
# Configuration : Registre des figures
import matplotlib.pyplot as plt
from datetime import datetime
from IPython.display import HTML, display

# Dictionnaire pour stocker les figures
figures_registry = {}

def register_figure(fig_id, title, description=""):
    """Enregistre une figure dans le registre global"""
    figures_registry[fig_id] = {
        'title': title,
        'description': description,
        'id': fig_id
    }

def generate_table_of_figures():
    """Génère la table des figures en HTML"""
    if not figures_registry:
        return HTML("<p><em>Aucune figure enregistrée</em></p>")
    
    html = "<div style='border: 1px solid #ddd; padding: 15px; background: #f9f9f9; border-radius: 5px;'>"
    html += "<h3>📊 Table des Figures</h3>"
    html += "<ol>"
    
    for fig_id, fig_data in figures_registry.items():
        html += f"<li><a href='#{fig_id}'><strong>{fig_data['title']}</strong></a>"
        if fig_data['description']:
            html += f" - <em>{fig_data['description']}</em>"
        html += "</li>"
    
    html += "</ol></div>"
    return HTML(html)

# Afficher la table (mise à jour automatiquement au fur et à mesure)
print("✅ Système de table des figures activé!")
```

## Exemple d'utilisation

Voici comment créer une figure qui apparaîtra automatiquement dans la table :

```python
# Exemple 1 : Figure avec ancre
fig, ax = plt.subplots(figsize=(10, 6))
ax.plot([1, 2, 3, 4], [1, 4, 2, 3], marker='o', linewidth=2)
ax.set_title('Impact d\'un événement dynamique (bouchon) sur le temps de trajet')
ax.set_xlabel('Temps (heures)')
ax.set_ylabel('Durée (minutes)')

# Ajouter l'ancre HTML
display(HTML("<a id='fig1'></a>"))

# Enregistrer dans la table
register_figure('fig1', 'Impact d\'un événement dynamique (bouchon)', 'Analyse temporelle')

plt.tight_layout()
plt.show()
```

```python
# Exemple 2 : Deuxième figure
fig, ax = plt.subplots(figsize=(10, 6))
algorithms = ['Glouton', 'Fourmis', 'Aléatoire']
temps = [2.5, 0.8, 5.2]
colors = ['#ff9999', '#66b3ff', '#99ff99']

ax.bar(algorithms, temps, color=colors)
ax.set_ylabel('Temps d\'exécution (secondes)')
ax.set_title('Comparaison des temps d\'exécution')

display(HTML("<a id='fig2'></a>"))
register_figure('fig2', 'Comparaison des temps d\'exécution (Glouton vs Fourmis)', 'Performance')

plt.tight_layout()
plt.show()
```
